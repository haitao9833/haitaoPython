## 二维
>- 可视化方便

### 读取数据

```python
data = read_ex(r'data\客户信息.xlsx')
data.head()
```

```python
allCol(data)
```

### 初始分布

```python
draw_Scatter2D(
    data['年龄(岁)'] , 
    data['收入(万元)'] ,
    Xscale = 5 , 
    Yscale = 5 , 
    c = False , 
    title = '客户信息'
).render_notebook()
```

### 训练模型 

```python
from sklearn.cluster import KMeans
```

```python
# n_clusters 即 K 值，即指定将所有样本分为几个类别
# random_state 指定随机选取初始中心点的种子

model = KMeans(n_clusters = 5 , random_state = 529833)
model.fit(data)
```

```python
# 分类结果标签

label = model.labels_
label[:10]
```

```python
data['分类'] = label
allCol(data)
data.head()
```

## 多维

```python
data = read_ex(r'data\肿瘤数据.xlsx')
allCol(data)
data.head()
```

```python
model = KMeans(n_clusters = 2 , random_state = 529833)
model.fit(data.drop(columns = '肿瘤性质'))
```

```python
# 分类结果标签

label = model.labels_
label[:10]
```

```python
data['分类'] = label
allCol(data)
data.head()
```

```python
from sklearn.metrics import accuracy_score as ACC
```

```python
# 分类 != 预测

ACC(data['肿瘤性质'] , data['分类'])
```

# `DBSCAN`
>- [可视化]https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/

## 读取数据

```python
data = read_ex(r'data\演示数据.xlsx')
allCol(data)
data.head()
```

## 初始分布

```python
draw_Scatter2D(
    data['x'] , 
    data['y'] ,
    Xscale = 0.5 , 
    Yscale = 0.5 , 
    c = False , 
    title = '演示数据'
).render_notebook()
```

## 训练模型

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(
    eps = 0.5 ,            # 画圆半径
    min_samples = 5        # 圆内样本数阈值
)
model.fit(data)
```

```python
# 分类结果标签

data['分类'] = model.labels_
allCol(data)
data.head()
```

```python
val_cnt(data , '分类')
```

## 可视化分布

```python
draw_Scatter2D_clf(
    data , 
    'x' , 
    'y' , 
    '分类' , 
    Xscale = 0.5 , 
    Yscale = 0.5
).render_notebook()
```

## 与 `KMeans` 对比

```python
scatter_pre , scatter_after =  model_Cluster(
    data , 
    2 , 10 , 
    0.5 , 0.5 , 
    '一条龙'
)
```
