# 先分类再降维测试

```python
# 读取数据

data = read_ex(r'data\肿瘤数据.xlsx').drop(columns = '肿瘤性质')
allCol(data)
data.head()
```

```python
# 先按所有特征聚类

model = KMeans(n_clusters = 5 , random_state = 529833)
model.fit(data)

data['分类'] = model.labels_
allCol(data)
data.head()
```

```python
val_cnt(data , '分类')
```

```python
# 保存分类

labelsInt = [int(la) for la in list(data['分类'])]
```

```python
# 消除量级

data = standardize_mean_var(data).drop(columns = '分类')
allCol(data)
data.head()
```

```python
# 降维

pca = PCA(n_components = 3)
pca.fit(data)
res3D = pca.transform(data)

print(type(res3D))
print(res3D.shape)
res3D[:10]
```

```python
pca_func(pca)
```

```python
data = pd.DataFrame(res3D , columns = ['一' , '二' , '三'])
allCol(data)
data.head()
```

```python
# 绘制

XYZ = [
    (
        float(data['一'].iloc[i]) , 
        float(data['二'].iloc[i]) , 
        float(data['三'].iloc[i])
    )
    for i in range(len(data))
]

LABELS_DICT_Int = {
    0 : '#abd9e9' , 
    1 : 'purple' , 
    2 : '#fdae61' ,
    3 : '#f46d43' , 
    4 : '#4575b4' , 
}
```

```python
draw_Scatter3D_clf(
    XYZ , 
    labelsInt , LABELS_DICT_Int , 
).render_notebook()
```
